```python
# Here we try to get the most watched movies (Total Views / Total Users Views / Total watch time)
# For series we concatenated the Session episode to differentiate between episodes

grouped=df.copy()
grouped.loc[grouped['program_class'] == 'SERIES/EPISODES', 'program_name'] = grouped['program_name']+'_SE'+grouped['season'].astype(str)+'_EP'+grouped['episode'].astype(str)
grouped = grouped.groupby(['program_name','program_class'])\
.agg({'user_id_maped': [('co1', 'nunique'),('co2', 'count')],\
      'duration_seconds': [('co3', 'sum')] }).reset_index()
grouped.columns = ['program_name','program_class','No of Users who Watched', 'No of watches', 'Total watch time in seconds']
grouped['Total watch time in houres']=grouped['Total watch time in seconds']/3600
grouped = grouped.drop(columns=['Total watch time in seconds'])
grouped = grouped.sort_values(by=['Total watch time in houres', 'No of watches','No of Users who Watched'], ascending=False).reset_index(drop=True)
```

# Customer Experience Study:

```python
# Here we try to study the customer experience against Program class
grouped=df.copy()
grouped = grouped.groupby('program_class')\
.agg({'user_id_maped': [('co1', 'nunique'),('co2', 'count')],\
      'duration_seconds': [('co3', 'sum')] }).reset_index()
grouped.columns = ['program_class','No of Users who Watched', 'No of watches', 'Total watch time in seconds']
grouped['Total watch time in houres']=grouped['Total watch time in seconds']/3600
grouped = grouped.drop(columns=['Total watch time in seconds'])
grouped = grouped.sort_values(by=['Total watch time in houres', 'No of watches','No of Users who Watched'], ascending=False).reset_index(drop=True)
```

```python
# show the result
grouped.head()
```

```python
# plot the total watch time against total number of users and report your findings
fig = make_subplots(rows=1, cols=2, specs=[[{'type':'pie'}, {'type':'pie'}]])

fig.add_trace(go.Pie(labels=grouped['program_class'], values=grouped['Total watch time in houres'], name='Total Watch Time'),1, 1)

fig.add_trace(go.Pie(labels=grouped['program_class'], values=grouped['No of Users who Watched'], name='Number of Users'),1, 2)

fig.update_layout(title='Program Analysis by Type')
fig.show()
```

```python
fig = px.pie(grouped, values='No of watches', names='program_class',\
             hover_data=['program_class'],title='No. of watches by program_class')

fig.update_traces(sort=False)
fig.show()
```

```python
# Reshape the data for better visualization
melted_grouped = grouped.melt(id_vars= 'program_class', var_name= 'Metric', value_name= 'Value')
```

```python
fig = px.bar(melted_grouped, x='program_class', y='Value', color='Metric', barmode='group', title='Comparison of program classes')
fig.show()
```

**Findings:**



From previous charts we can noticed that series class has a higher number of watches and a longer total watching time compared to movies.

> This means that series have many prolonged watching sessions, which may indicate continued demand and great interest in watching the episodes.

While the Movies class has more users who watched it, but less overall time than series.

> This may indicate that movies attract a wider audience but for one-time or short viewing periods, which reflects the nature of movies which are usually one complete watching rather than over a series of episodes like soap operas.
