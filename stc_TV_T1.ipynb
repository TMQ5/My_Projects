# Studying the Relationship of User Watches (watch resolution HD or SD):

```python
"""
TODO study the relation and user's behaviour Against HD flag, feel free to include any comparison
"""
resolution = df.copy()
resolution.loc[resolution['program_class'] == 'SERIES/EPISODES', 'program_name'] = resolution['program_name']+'_SE'+resolution['season'].astype(str)+'_EP'+resolution['episode'].astype(str)
resolution_grouped = resolution.groupby(['program_class', 'program_name'])\
  .agg({'user_id_maped': 'nunique',
        'hd': ['count', lambda x: (x==1).sum(), lambda x: (x==0).sum()],
        'duration_seconds': 'sum'}).reset_index()

resolution_grouped.columns = ['_'.join(col).strip() if col[1] else col[0] for col in resolution_grouped.columns.values]

resolution_grouped['Total watch time in hours'] = resolution_grouped['duration_seconds_sum'] / 3600
resolution_grouped.drop(columns=['duration_seconds_sum'], inplace=True)

columns = ['program_class', 'program_name', 'user_id_maped_nunique', 'hd_count', 'hd_<lambda_0>', 'hd_<lambda_1>', 'Total watch time in hours']
resolution_final = resolution_grouped[columns]

resolution_final.columns = ['Program Class', 'Program Name', 'No of Users who Watched', 'Total No of Watches', 'HD Watches', 'SD Watches', 'Total Watch Time in Hours']
```

```python
# Show the result
resolution_final.head(10)
```

```python
# Plot a pie chart to visualize the distribution of watch quality based on the total number of watches for HD and SD.
values = resolution_final[['HD Watches', 'SD Watches']].sum().tolist()
names = ['HD', 'SD']

fig = px.pie(values=values, names=names, title='Distribution of Watch Quality')
fig.show()
```

The previous chart indicates that the number of HD watches quality is 405,036, representing 38.6% of the total watches, while the number of SD quality watches is 643,539, representing 61.4% of the total watches.

This precentage shows a preference for SD quality over HD, as there is a 58.88% difference between the number of SD watches and HD wathces, equivalent to 238,503 watches.

This significant difference suggests a user preference for this quality or cause to possibly other factors that prevent users from utilizing HD quality, such as internet connection quality constraints, the type of devices used that may not support HD quality, or data costs.

Based on this information, it might be advisable for STC to improve the user experience for the STC TV application and provide content that enhances the watching experience in SD quality.

```python
# Creating a heatmap to displays the correlation matrix of numerical data in 'resolution_final' DataFrame.
# Selecting only numeric data types from the DataFrame for correlation analysis.
numeric_df = resolution_final.select_dtypes(include=[np.number])

# Calculating the correlation matrix for the numeric columns in the DataFrame.
correlation_matrix = numeric_df.corr()


fig = px.imshow(correlation_matrix,
                labels=dict(x="Variables", y="Variables", color="Correlation Coefficient"),
                x=correlation_matrix.columns,
                y=correlation_matrix.columns,
                text_auto=True,
                title="Heatmap of Correlation Matrix")

fig.update_xaxes(side="bottom")

fig.show()
```

We can assert that there is a strong positive relationship between the number of users who watch programs on the STC TV app and the number of watches in HD quality; that is, as the number of users within the app who choose HD quality increases, so does the number of watches in this quality. Similarly, there is also a strong positive relationship in terms of the number of watches: as the number of watches for a particular program on the STC TV app increases, so does the number of watches that occur in HD quality (this indicates that the more popular programs tend to attract more watches in high quality).

As for the number of hours that app users spend watching programs, and the number of watches in both HD and SD qualities, there is a moderately positive relationship between them, although the number of watches in HD quality is slightly higher. This suggests that although watches in HD quality are not the only ones that affect the duration of watching, they have a noticeable impact.

```python
# Aggregating the total number of HD and SD watches by program class from the 'resolution_final' DataFrame.
grouped = resolution_final.groupby('Program Class')\
    .agg({'HD Watches': 'sum',
          'SD Watches': 'sum'}).reset_index()

# Creating a stacked bar chart to display the sums of HD and SD watches by program class.
fig = px.bar(grouped, x='Program Class', y=['HD Watches', 'SD Watches'],
             title= 'Watch Quality by Program Class',
             labels= {'value': 'Number of Watch', 'variable': 'Qulaity Type'})

fig.show()
```

The chart above shows that movies in HD quality attract more watches than series, while series in SD quality draw more wathces than movies. The high production quality of movies and the complex visual details they contain may enhance viewers' preference to watch them in HD, leading to an increase in the number of watches in this category with higher quality.

On the other hand, series are characterized by their longer duration and frequent watching, making them more suitable for watching in SD quality, especially in conditions of limited data or internet connectivity.
